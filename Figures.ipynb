# Plotting Functions

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import cm 
%matplotlib inline


def plot_figure(dataset1,dataset2,name,colour,lw,ax):
    ax.scatter(x,dataset1,marker='_',c=colour, s=300,linewidth=1)
    ax.scatter(x,dataset2,marker='_',c=colour, s=300,linewidth=1)
    ax.plot(x,dataset1,linestyle=':',color=colour,linewidth=lw, label=name)
    ax.plot(x,dataset2,linestyle=':',color=colour,linewidth=lw)
    
    
def plot_style_A():
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_linewidth(1.5)
    ax.spines['bottom'].set_visible(False)
    plt.tick_params(axis='x', top='off')
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_tick_params(labelsize=12,width=1.5)
    ax.yaxis.set_tick_params(labelsize=12,width=1.5)
    plt.ylabel('Relative Energy (eV)',fontsize=12)
    plt.tight_layout()

def xlabels(labelnames):
    labels = [item.get_text() for item in ax.get_xticklabels()]
    labels = [i for i in labelnames]
    plt.xticks(x, labels,rotation=30,fontsize=10)
def legend_style_horizontal():
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.05,
                 box.width, box.height * 0.95])

    ax.legend(loc='upper center', bbox_to_anchor=(0.2, +1.),
            ncol=2,fontsize=10,markerscale=0.75, frameon=False)

def legend_style_vertical():
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.width*0.05,
                 box.width*0.95, box.height])

    ax.legend(loc='best', bbox_to_anchor=(0.2, 1),
            ncol=1,fontsize=10,markerscale=0.75, frameon=False)
```

# Plots Comparing Models

```python
def plot_figure(dataset1,dataset2,name,colour,lw,ax):
    ax.scatter(x,dataset1,marker='_',c=colour, s=100,linewidth=1)
    ax.scatter(x,dataset2,marker='_',c=colour, s=100,linewidth=1)
    ax.plot(x,dataset1,linestyle=':',color=colour,linewidth=lw, label=name)
    ax.plot(x,dataset2,linestyle=':',color=colour,linewidth=lw)

def legend_style_horizontal(ax):
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.05,
                 box.width, box.height * 0.95])
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, +0.15),
            ncol=2,fontsize=8,markerscale=0.2, frameon=True)    

fig, ((ax0,ax1),(ax2,ax3))=plt.subplots(ncols=2,nrows=2,sharey=True)
x=range(len(H_lM_ON_s0))
plot_figure(H_lM_ON_s0,H_lM_ON_s1,"1-lM(EE)","red",1,ax0)
plot_figure(Hmech_lM_ON_s0,Hmech_lM_ON_s1,"1-lM(ME)","darkred",2,ax0)
ax0.set_title("Electrostatic vs Mechanical Embedding",fontsize=8)
#x=[0,2,3]
#plot_figure(pMeO_lM_ON_s0,pMeO_lM_ON_s1,"2-lM(EE)","black",1,ax0)
#plot_figure(pMeOmech_lM_ON_s0,pMeOmech_lM_ON_s1,"1-lM(ME)","grey",2,ax0)
legend_style_horizontal(ax0)
plt.tight_layout()

x=range(len(H_lM_ON_s0))
labelnames=["FC","E*min","K*min","K*MECI"]

for ax in [ax0,ax1,ax2,ax3]:

    labels = [item.get_text() for item in ax.get_xticklabels()]
    labels = [i for i in labelnames]
    ax.set_xticks(x)
    ax.set_yticks(range(6))
    ax.set_xticklabels(labels,rotation=30)
    ax.xaxis.set_tick_params(labelsize=10,width=1.5)
    ax.yaxis.set_tick_params(labelsize=10,width=1.5)
    if ax==ax0 or ax==ax2:
        ax.set_ylabel('Relative Energy (eV)',fontsize=10)
plt.tight_layout()

```
